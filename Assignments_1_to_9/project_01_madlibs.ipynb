```python
import random
import os
from datetime import datetime
from colorama import init, Fore, Style

# Initialize Colorama
init(autoreset=True)

def welcome():
    print(Fore.CYAN + Style.BRIGHT + "ğŸ‰ Welcome to the ULTIMATE MAD LIBS PRO EDITION! ğŸ‰")
    print(Fore.YELLOW + "ğŸ“– Fill in the blanks and create hilarious, wild, or weird stories!")
    print("ğŸ¤– Let's unleash your creativity!\n")

def get_user_inputs(placeholders):
    print(Fore.GREEN + "ğŸ§  Please provide the following words:")
    user_inputs = {}
    for word_type in placeholders:
        value = input(f"âœï¸ Enter a {word_type}: ")
        user_inputs[word_type] = value
    return user_inputs

def apply_template(template, user_inputs):
    story = template
    for key, value in user_inputs.items():
        story = story.replace(f"<{key}>", value)
    return story

def save_story(story_text):
    filename = f"madlibs_{datetime.now().strftime('%Y%m%d_%H%M%S')}.txt"
    with open(filename, "w", encoding="utf-8") as file:
        file.write(story_text)
    print(Fore.MAGENTA + f"\nğŸ’¾ Your story has been saved as: {filename}!\n")

def choose_template():
    templates = [
    {
        "text": "ğŸŒ™ Last night, I dreamed I was a <adjective> <noun> who loved to <verb> at the <place> ğŸŒŒ.",
        "placeholders": ["adjective", "noun", "verb", "place"]
    },
    {
        "text": "ğŸ§³ Today, my <relative> and I went to the <place>. We saw a <adjective> <animal> that <verb> past us! ğŸ¾",
        "placeholders": ["relative", "place", "adjective", "animal", "verb"]
    },
    {
        "text": "ğŸ° Once upon a time in <place>, there lived a <adjective> <profession> named <name>. They always <verb> on Sundays. ğŸ“†",
        "placeholders": ["place", "adjective", "profession", "name", "verb"]
    },
    {
        "text": "ğŸš€ On planet <place>, a <adjective> alien named <name> loved to <verb> with their pet <animal> ğŸ›¸.",
        "placeholders": ["place", "adjective", "name", "verb", "animal"]
    },
    {
        "text": "ğŸ“ My favorite teacher, <name>, always said '<verb> like a <adjective> <animal> in the <place>!' ğŸ«",
        "placeholders": ["name", "verb", "adjective", "animal", "place"]
    },
    {
        "text": "ğŸ• I ordered a <adjective> pizza topped with <noun>, <animal> sprinkles, and a dash of <place>. It tasted like <verb>! ğŸ¤¢",
        "placeholders": ["adjective", "noun", "animal", "place", "verb"]
    },
    {
        "text": "ğŸ® In the game of life, you are the <adjective> <profession> who must <verb> the evil <animal> boss in <place>! ğŸ’¥",
        "placeholders": ["adjective", "profession", "verb", "animal", "place"]
    },
    {
        "text": "ğŸ¤ Breaking News! <name> the <adjective> <profession> just <verb> live on TV in front of the entire <place>! ğŸ“º",
        "placeholders": ["name", "adjective", "profession", "verb", "place"]
    },
    {
        "text": "ğŸ§™ A <adjective> wizard named <name> cast a spell to <verb> every <animal> in <place> into dancing marshmallows. ğŸ¬",
        "placeholders": ["adjective", "name", "verb", "animal", "place"]
    },
    {
        "text": "ğŸ¢ My <relative> and I rode a <adjective> rollercoaster that screamed '<verb>!' every time we passed a <animal> in <place>! ğŸ˜‚",
        "placeholders": ["relative", "adjective", "verb", "animal", "place"]
    }
]

    return random.choice(templates)

def display_story(story):
    print(Fore.CYAN + "\nğŸ“š Here's your Mad Lib Story:\n")
    print(Fore.LIGHTWHITE_EX + "-" * 60)
    print(story)
    print("-" * 60)

def ask_replay():
    choice = input(Fore.YELLOW + "\nğŸ” Want to create another story? (yes/no): ").strip().lower()
    return choice in ["yes", "y"]

def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

def play_game():
    while True:
        clear_screen()
        template_data = choose_template()
        user_inputs = get_user_inputs(template_data["placeholders"])
        story = apply_template(template_data["text"], user_inputs)
        display_story(story)
        save_story(story)
        if not ask_replay():
            print(Fore.CYAN + "\nğŸ‘‹ Thanks for playing the Ultimate Mad Libs Pro Edition!")
            print("âœ¨ Keep being creative & coding! ğŸ’»\n")
            break

def run():
    welcome()
    play_game()

if __name__ == "__main__":
    run()
```
